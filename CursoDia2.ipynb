#IF



```python
# Como insertar las condicionales en un if
if 5>4:
  print("5 si es mayor que 4")
# Pregunta que: si 5 es mayor que 4, entonces se ejecute lo de adentro.

# Ejercicio de compra de tubos
costoTuboNormal = 20
costoDecapadoZinc = 10
decapadoZinc = True

# si el cliente decado de zinc se le cobran 10 adicionales al costo del tubo ()
pagoCliente = costoTuboNormal
if decapadoZinc :
  pagoCliente = costoTuboNormal + costoDecapadoZinc
print(pagoCliente)

```

# IF/ELSE

```python
## calificaciones aprobatorias
calificacion = 80.2
if calificacion >= 70: # si es mayor o igual a 70
  print("Alumno Aprobado")
else:
  print("Alumno Reprobado")

# clasificacion de las edades
edad = int(input("Ingresa tu edad: "))
if edad >=18:
  print("eres mayor de edad")
  if edad >= 65:
    print("Eres un adulto mayor")
  else:
    print("Eres un adulto")
else:
  print("Eres un menor de edad")
  if edad<=12:
    print("Eres un niño")
  else:
    print("Eres un adolescente")

calificacionUSA = ""
# calificaciones USA
if calificacion == 10:
  calificacionUSA = "A+"
elif calificacion >= 90:
  calificacionUSA = "A"
elif calificacion >= 80:
  calificacionUSA = "B"
elif calificacion >= 70:
  calificacionUSA = "C"
elif calificacion >= 60:
  calificacionUSA = "D"
else:
  calificacionUSA = "F"

print("Tu calificacion es de " + calificacionUSA)

# Saber si el numero es par o impar
numero = int(input("Ingresa un numero: "))
if numero % 2 == 0:
  print("Tu numero es par")
else:
  print("Tu numero es impar")

```

# clicos for

```python
# imprimir los numeros del 0 al 4
for i in range(5): # empieza en 0 y RECORRE 5 posiciones (hasta el cuatro)
  print(i)
print("----------")
for i in range(2,10): #rango con un inicio y un final, sin incluir el final [2,10)
  print(i)

print("----------")
for i in range(3,31,3): # empieza en 3, termina en 30 y va de 3 en 3
  print(i)
```

# CLICLO WHILE

```python
suma = 0.0
num = float(input("Inserta un numero para sumar o 0 para salir: "))
while num!=0: # mientras num sea diferente a cero será verdadero
  suma += num
  num = float(input("Inserta un numero para sumar o 0 para salir: "))
print(f"Tu suma es: {suma}")
```
