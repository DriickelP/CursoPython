# Listas
permiten tener ordenados los elementos y almacenan elementos de diferentes tipos de datos



```python
# Posiciones:      0         1         2         3        4
listaCompras = ["huevo", "Tomate", "Cebolla", "Chile", "Jamon"]
print(listaCompras) # imprime toda la lista
print(listaCompras[4]) # imprime la posicion numero 4

## Insertar elementos en la lista
#Append: la pone al final
listaCompras.append("Cereal")
print(listaCompras)

#Insert: inserta en una posicion
listaCompras.insert(1,"Leche") # en la posicion 1 inserta "leche"
print(listaCompras)

```

```python
## eliminar elementos de una lista
# remove: elimina un elemento en concreto
listaCompras.remove("Cereal")

# pop: elimina el ultimo elemento
listaCompras.pop()
listaCompras.pop(2) # elimina la posicion 2

```

```python
## CICLOS con listas
for i in range(len(listaCompras)): # recorre segun la cantidad de elementos
  print(f"{i+1}-. {listaCompras[i]}") # hace una lista de compras en el indice del for

```

```python
promedios = [70.2, 85.2, 65.4, 89.5, 90.5, 71.6]
promedios.sort() # acomoda la lista de orden ascendente
print(promedios)

promedios.sort(reverse = True)
print(promedios)

# imprimir el ultimo elemento y los maximos y minimos
print(promedios[-1]) # ultimo elemento de la lista
print(max(promedios))
print(min(promedios))

calificaciones = [[70.2, 85.2, 65.4], [89.5, 90.5, 71.6]]
i = 1
for X in calificaciones:
  print(f"Alumno {i}:")
  j = 1
  for Y in X:
    print(f"     Calificacion {j}: {Y}")
    j += 1
  j=0
  i+=1
```

## SETS

```python
# pueden tener diferentes tipos de datos
setDatos = {1,2,3,4,5, "Hola", 1.2, True}
print(setDatos)

#los sets no admiten agregar en cierta posicion o leer cierta posicion
setDatos.add("juan") # si
print(setDatos[2]) #no
```

### Uniones de sets


```python
lista1 = {1,2,3,4,5}
lista2 = {5,6,7,8,9}
print("Listas originales")
print(lista1)
print(lista2)

listas1_2 = lista1.union(lista2) # o bien lista1 | lista2
print("\nEsto es una union")
print(listas1_2)

# une ambos sets, y sin estan repetidos coloca solo 1 de los dos


listasInt1_2 = lista1.intersection(lista2) # o bien lista1 & lista2
print("\nEsto es una interseccion")
print(listasInt1_2)
# Coloca los elementos que la listas comparten entre si.

listasDif1_2 = lista1.difference(lista2) # o bien lista1 - lista2
print("\nEsto es una Diferencia")
print(listasDif1_2)
# Resta los elementos que comparten entre s√≠
```
