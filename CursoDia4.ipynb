## Funciones
son secciones de codigo que podemos usar cuando nosotros queramos, como lo sol las que hemos utilizado. Para definir se utiliza el def

```python
def holaMundo():
  print("\nHola mundo") # Imprime directamente el mensaje en la consola
holaMundo()

def retornaHola():
  return "\nHola mundo con retorno" # Retorna un mensaje sin imprimirlo
print(retornaHola()) # Lo imrpime

def saludar(nombre):
  print(f"Hola {nombre} ¿Como estas?")
tuNombre = input("\nInserta tu nombre: ")
saludar(tuNombre)
```

```python
def sumatoria(num1, num2):
  return num1 + num2

suma = sumatoria(3,4)
print(suma)

def bienvenida(invitado = "Amigo"): # por defecto se mostrará amigo, a menos que se indique lo contrario
  print(f"Bienvenido, {invitado}")

bienvenida()
bienvenida("uriel")

def persona(edad, nombre="X"):
  return edad, nombre
cliente = persona(18, "Uriel")
print(f"Se retorno edad y nombre en tipo {type(cliente)}")
```

## Funciones que llaman a otras funciones

```python
# Heladeria ejercicio de heladeria con mi logica
def saboresHelados():
  sabores = ["vainilla", "fresa", "chocolate", "napolitano"]
  print("Sabores disponibles: ")
  for i in range(len(sabores)):
    print(f"- {sabores[i]}")
  eleccionSabor = input("Selecciona tu sabor: ").strip().lower()
  while eleccionSabor not in sabores:
    print("Aun no tenemos ese sabor.")
    eleccionSabor = input("Selecciona tu sabor: ").strip().lower()
  return eleccionSabor

def cantidadHelados(sabor):
  cantidad = int(input(f"Selecciona la cantidad de helados sabor {sabor}: "))
  while cantidad <= 0:
    cantidad = int(input("Selecciona una cantidad valida: "))
  return cantidad, sabor

def mostrador():
  tuplaDatos = cantidadHelados(saboresHelados()) # una funcion llama a 2 funciones, y dentro de una funcion se retorna un valor
  if len(tuplaDatos) == 2:
    print("Pedido Realizado")
    if tuplaDatos[0] == 1:
      print(f"Se pidió una pieza de helado de {tuplaDatos[1]}")
    else:
      print(f"Se pidió {tuplaDatos[0]} piezas de helado de {tuplaDatos[1]}")
mostrador()

```

## Funciones lambda
funcion acortada para escribir una funcion pero de forma diferente
nombreFuncion = lambda parametros : accion
la diferencia respecto a las otras es que son un poco mas limitadas


```python
sumar = lambda x,y : x+y
print(sumar(10,5))

descuentoLambda = lambda descuento, precio : (precio * (1- (descuento/100)), precio * (descuento/100))
precio = float(input("Coloca el precio: "))
descuento = float(input("Coloca el descuento en porcentaje: "))
tuplaPrecios = descuentoLambda(descuento, precio)

print(f"El precio {tuplaPrecios[0]} es con el {descuento} de {precio}, ({tuplaPrecios[1]})")
```

## Manejo de errores
nos ayuda a que el programa no se pare al momento de que ocurra un error y se pueda manejar ese error para ayudar al usuario

```python
# print(10/0) marcará error pues no se puede dividir por zero, por lo cual se utiliza la estructura
try:
  numeroA = float(input("Inserta el dividendo: "))
  numeroB = float(input("Inserta el divisor: "))
  resultado = numeroA/numeroB
except ZeroDivisionError:
  print("No puedes dividir por cero. Intentalo de nuevo")
```

```python
from typing_extensions import Doc
# Ejercicio: calculadora con manejo de errores
def pedirNumeros(caso):
  match caso:
    case 1:
      print("\nSuma")
    case 2:
      print("\nResta")
    case 3:
      print("\nMultiplicacion")
    case 4:
      print("\nDivision")
    case 5:
      print("\nPotencia")

  a = float(input("Inserta el primer numero: "))
  b = float(input("Inserta el segundo numero: "))
  return a,b

sumatoria = lambda tuplaNumeros : (tuplaNumeros[0] + tuplaNumeros[1])
resta = lambda tuplaNumeros : (tuplaNumeros[0] - tuplaNumeros[1])
multiplicacion = lambda tuplaNumeros : (tuplaNumeros[0] * tuplaNumeros[1])
division = lambda tuplaNumeros : (tuplaNumeros[0] / tuplaNumeros[1])
potencia = lambda tuplaNumeros : (tuplaNumeros[0] ** tuplaNumeros[1])

def menu():
  print("\nCalculadora\n1-. Suma\n2-. Resta\n3-. Multiplicacion\n4-. Division\n5-. Potencia")

menu()
opcion = int(input("Selecciona la accion a realizar: "))
while opcion != 6:

  try:
    match opcion:
      case 1:
        print(f"El resultado es {sumatoria(pedirNumeros(1))}")
      case 2:
        print(f"El resultado es {resta(pedirNumeros(2))}")
      case 3:
        print(f"El resultado es {multiplicacion(pedirNumeros(3))}")
      case 4:
        print(f"El resultado es {division(pedirNumeros(4))}")
      case 5:
        print(f"El resultado es {potencia(pedirNumeros(5))}")
  except ValueError:
    print("Error. Por favor inserta numeros")
  except ZeroDivisionError:
    print("Error. No se puede dividir por cero")
  menu()
  opcion = int(input("Selecciona la accion a realizar: "))
```
