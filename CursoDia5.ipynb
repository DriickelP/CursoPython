## Modulos Integrados


```python
# Existen librerias que ya estan por defecto en python utilizando:
import math
raiz = math.sqrt(2)
print(f"{raiz:.4f}") # :.4f se refiere a que solo 4 decimales
```

```python
# De igual modo se pueden poner librerias con un alias
import math as mates ## math ahora se llama mates
root = mates.sqrt(8)
print(f"{root:.2f}")
```

## Random
nos permite hacer diferentes funciones pseudo random tales como generacion de numeros o elecciones.

```python
import random
numerosRandom = [random.randint(1,100) for _ in range(5)] ## creame una lista con 5 elementos random, el _ significa que esa variable no se va a usar
print(numerosRandom)
```

```python
import random
import string

caracteres = string.ascii_letters + string.digits + string.punctuation
password = "".join(random.choice(caracteres) for _ in range(12))
print(password)
```

## Libreria datetime
nos permite obtener datos relacionados con el tiempo, o con las fechas


```python
from datetime import datetime, timedelta
fechaActual = datetime.now() # fecha actual en UTC
# print(fechaActual)
fechaZonaHoraria = fechaActual - timedelta(hours=6) # me mostraba la hora UTC, por lo cual le reste 6 horas (UTC-6 en mexico)
print(fechaZonaHoraria.strftime("%d/%m/%Y %H:%M:%S")) # Le da formato a la fecha qeu se tiene
```

## Formato JSON
es un formato muy utilizado para guardar y transmitir datos son muy utilizados en APIS
Es muy parecido a los diccionarios, mas sin embargo es solo texto
- Sirve para convertir objetos de python a texto JSON y tambien de texto JSON a objetos de python

Se le llama **Serializar** a:
- Convertir un objeto de python en texto JSON

Se le llama **Desereallziar** a:
- Convertir un texto JSON a un objeto usable en python


```python
## Funciones Principales
#1) json.dumps(obj): convierte diccionario u objeto de python en un string de JSON
import json
obj = {
    "nombre" : "Uriel Pesina",
    "edad" : 25}
print(obj)
print(type(obj))
s = json.dumps(obj, ensure_ascii=False)
print(s)
print(type(s))
```

```python
##2) json.loads(texto_json): convierte un string JSON en diccionario python
restauracion = json.loads(s)
print(s)
print(type(s))
```

## Numpy
provee un tipo de array , operaciones vectorizadas, no es igual a las listas de python pero solo tiene que tener datos de un solo tipo


```python
#pip install numpy
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) # Arreglo (2,3)
V = np.array([10,20,30])

print(A+V)

prueba = [1,2,3]
prueba2 = [1,2,3]
print(f"\n{prueba + prueba2}") # No los suma, solo los concatena
```

```python
zero = np.zeros((2,3))
uno = np.ones((2,3))
matrizRandom = np.random.rand(2,3)
print(zero)
print(uno)
print(matrizRandom)
```

```python
nums = np.arange(0,10,2) # Saca valores entre 0 y 10 pero de 2 en dos
print(nums)

x = np.linspace(0,1,5) # saca 5 valores de entre 0 y 1, con igual separacion entre si
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
```

```python
print(a.dot(b))

# multiplicacion de matrices (producto cruz)
c = np.array([[1,2],[3,4],[5,6]])
# a b
# c d
# e f

d = np.array([[1,2,3],[4,5,6]])
# g h i
# j k l

print(c.dot(d))
```

```python
datos = np.array([1,2,3,4,5])
print(datos.min()) # minimo
print(datos.max()) # maximo
print(datos.sum()) # suma
print(datos.mean())# media
print(datos.std()) # desviacion estandar

A = np.array([1,6,3,9,2])
print(A[A>5]) # De el arreglo A solo traeme los valores mayores a 5
print(A[A<5]) # De el arreglo A solo traeme los valores menores a 5
```

## Mtplotlib

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,3,5,7,11]
# plt.plot(x,y)
plt.plot(x,y,marker="o") # coloca con una O las intersecciones
plt.title("Titulo de la grafica")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.grid() # coloca una cuadricula
plt.show()

# nos permite realizar graficas con base de datos
```
