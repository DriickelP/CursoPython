# Metodo: de instancia y de clase
un metodo de instancia es una funcion definida dentro de una clase que

- Puede acceder y modificar los atributos de ese objeto

- Recibe como primer parametro a self

```python
class tuClase:
  def metodo(self):
    pass
```

```python
class Calculadora:
  def __init__(self,A,B):
    self.A = A
    self.B = B
  def sumar(self):
    print(self.A + self.B)
  def resta(self):
    print(self.A - self.B)
  def multiplicacion(self):
    print(self.A * self.B)
  def division(self):

    if self.B == 0:
      print("No puedes dividir por cero")
    else:
      print(self.A / self.B)
Calculadora = Calculadora(7,8)
Calculadora.division()
Calculadora.sumar()
Calculadora.resta()
Calculadora.multiplicacion()
```

## 6-. encapsulamiento y propiedades

```python
class Cuentas:
  def __init__(self, saldo):
    self.__saldo = saldo #variable privada, proteje los datos de un objeto
cuenta = Cuentas(500)
cuenta.Cuentas = 1000
# print(cuentas.__saldo) marca error pues no se puede acceder directamente

```

```python
# encapsulamiento
# Getters y setters

class Cuentas:
  def __init__(self, saldo):
    self.__saldo = saldo
  @property
  def saldo(self): # Es un Getter
    # Permite leer el valor del saldo usando el metodo
    return self.__saldo # retorna la variable privada
  @saldo.setter
  def saldo(self, valor):
    self.__Saldo = valor

c = Cuentas(1000)
c.saldo
c.saldo = 2000
c.saldo # no cambió
```

7-. Herencia y abstraccion

```python
class Vehiculo: # Clase PADRE o clase BASE
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo
  def descripcion(self):
    return f"{self.marca}: {self.modelo}"
  def mover(self):
    return "El vehiculo se mueve"

#subclase
class Coche(Vehiculo): # clase hija
  def mover(self):
    return "El carro esta conduciendo"

class Moto(Vehiculo):
  def mover(self):
    return "La moto se mueve en 2 llantas"
coche2 = Vehiculo("BYD", "Dolphin")
coche = Coche("Toyota", "Corolla")
moto = Moto("Yamaha", "R6")
print(coche2.descripcion())
print(coche2.mover())

print(coche.descripcion()) #aun que no volvimos a declarar el metodo dentro de descripcion, podemos seguir usandolo ya que lo heredo del padre
print(coche.mover())

print(moto.descripcion())
print(moto.mover())
```

```python
class Persona:
  def __init__(self,nombre, edad):
    self.nombre = nombre
    self.edad = edad
  def presentarse(self):
    print(f"Hola, soy {self.nombre} y  tengo {self.edad} años")

class Estudiante(Persona):
  def __init__(self,nombre,edad,carrera):
    super().__init__(nombre,edad) # manda a llamar a las variables de la clase padre
    self.carrera = carrera
  def presentarse(self):
    super().presentarse() # manda a llamar a presentarse
    print(f"Estudio la carrera de: {self.carrera}")
# Reutiliza el constructor de la clase padre

est = Estudiante("Ana",21,"programacion")
est.presentarse()
```

## Composicion HAS-A
es una composicion de un objeto grande a aprtir de objetos, se compone de diversos objetos para hacer uno solo
- Una clase no hereda de otra, si no que contiene un objeto de esa otra clase


```python
class Banco:
  def __init__(self,dinero):
    self.dinero = dinero
  def verTotal(self):
    print(f"Total de la cuenta ${self.dinero}")
class Persona:
  def __init__(self, cuenta: Banco): # indica que esta variable es una instancia de la clase banco
    self.cuenta = cuenta # la persona tiene un objeto banco, NO LO HEREDA
```

## composicion vs herencia
ninguna es mejor que otra, simplemente hacen cosas diferentes
herencia: A es B; gato es un animal
composicion: A tiene como posesion a B: persona tiene cuenta de banco
