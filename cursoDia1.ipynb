# Operaciones Matematicas

```python
# Suma
print(3+5)
numeroUno, numeroDos = 10,20
print(f"la suma de {numeroUno} + {numeroDos} = {numeroUno + numeroDos}")

# Resta
print(10-7)
v1,v2 = 14.2, 16.8
resultado = v2-v1
print(resultado)

# Multiplicacion
M1,M2,M3,M4,M5 = 5.3,10.2,15.1,20.5,25.3
print(f"\n{M1} * {M2} = {M1*M2}")


# Division
print(f"\n{M3} / {M4}")

# Modulo "Residuo"
print(f"El residuo de: {M5} / {M1} es {M5%M1}")

# Potencia
print(f"{M1} elevado a la 3 es igual a {M1**3}")

# Area de un circulo (parentesis)
pi,radio = 3.1415926535, 0.3
area = (pi * (radio**2))/2
print(f"\nArea = {area}")
```

# Operadores Booleanos

```python
# No es mas que las compuertas logicas
# multiplicacion (and)
cero = False
uno = True
andFalsa = cero and uno # viceversa es igual
andVerdadera = uno and uno
print(f"-----\n{andFalsa} y {andVerdadera}")

# suma (or)
orFalsa = cero or cero
orVerdadera = cero or uno
orVerdadera1 = uno or uno
print(f"-----\n{orFalsa} | {orVerdadera} | {orVerdadera1}")

# inversa (not)
orFalsaInversa = not (cero or cero)
orVerdaderaInversa = not (cero or uno)
print(f"------\n{orFalsaInversa} y {orVerdaderaInversa}")

# quiero ir a una fiesta si va juan o pedro, pero no los dos por que se pelean (xor)
#(á*b) + (a*b´)
juan = True
pedro = True
#¿si van los dos podre ir?
fiesta = ((not(juan) and pedro) + (not(pedro) and juan))
print(fiesta)

#¿Si va juan podre ir?
juan = False
pedro = True
#¿si van los dos podre ir?
fiesta = ((not(juan) and pedro) + (not(pedro) and juan))
print(fiesta)
```
